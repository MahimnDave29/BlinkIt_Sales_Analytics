```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('BlinkIt.xlsx')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
item_weight_mean = df.groupby('Item_Identifier')['Item_Weight'].transform('mean')
df['Item_Weight'].fillna(item_weight_mean, inplace=True)
```

```python
df['Item_Weight'].fillna(item_weight_mean, inplace=True)
```

```python
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
item_visibility_mean = df[df['Item_Visibility'] > 0].groupby('Item_Type')['Item_Visibility'].mean()
df['Item_Visibility'] = df.apply(lambda row: item_visibility_mean[row['Item_Type']]
                                 if row['Item_Visibility'] == 0 else row['Item_Visibility'], axis=1)
```

```python
df['Outlet_Age'] = 2025 - df['Outlet_Establishment_Year']
df['Item_Category'] = df['Item_Identifier'].str[:2]
df['Price_Category'] = pd.cut(df['Item_MRP'], bins=[0, 100, 150, 200, 300], labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
df.head()
```

```python
df.to_csv('BlinkIt_Cleaned.csv', index=False)
```
